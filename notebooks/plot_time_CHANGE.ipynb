```python
def time2min(t: str) -> int:
    v = t.split(':')
    if len(v) == 1:
        return int(v)
    
    total = int(v[0]) * 60 + int(v[1]) + (1 if (int(v[2]) / 60. > 0.5) else 0)
    return total
```

```python
time2min('1:57:10')
```

```python
time_T = [4, 8, 12, 16, 20, 24, 28]
time_cost = [time2min('1:57:10'), 
             time2min('3:53:51'),
             time2min('5:05:14') + time2min('3:43:54'),
             time2min('5:03:30') + time2min('7:54:13'),
             time2min('4:18:33') + time2min('15:20:19'),
             time2min('5:08:08') + time2min('13:17:19'),
             time2min('3:50:59') + time2min('20:16:41'),
            ]
```
